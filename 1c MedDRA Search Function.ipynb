```python
import pandas as pd
import numpy as np
import re
import random
```

```python
# Importing NLTK

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
# Importing MedDRA Library
meddra = pd.read_csv('Datasets/llt.tsv', sep='\t')
meddra
```

```python
# Importng final result
ade_result = pd.read_csv('../JPM_SMM4H_1b_LW/bert-large/20_epochs_baseline/final_test_result/unformatted_merged_result.tsv', sep='\t')
ade_result
```

```python
best_results = []

for index, ade_row in ade_result.iterrows():
    predict_span = ade_row['predict_span']

    # pre-processing predict_span to get rid of stop words
    predict_span_words = [word for word in predict_span.split() if word.lower() not in stop_words]

    # Going through each word in the span and each term in the meddra library to see if there is a match
    span_results = [] 
    for ind_word in predict_span_words:
        for index, meddra_row in meddra.iterrows():
            pt_term = meddra_row['PT']
            lt_term = meddra_row['LLT']
            word_term = meddra_row['Label']

            span_match = re.search(r"\b%s\b"%ind_word, word_term)
            
            if span_match != None:
                span_results.append((pt_term, word_term, lt_term))
            
    best_results.append(span_results)

    
```
